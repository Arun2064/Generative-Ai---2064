```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

file_path = '/content/Housing.csv'
df = pd.read_csv(file_path)

df = pd.get_dummies(df, drop_first=True)

X = df.drop('price', axis=1)
y = df['price']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = Sequential([
    Dense(18, activation='swish', input_shape=(X_train.shape[1],)),
    Dense(26, activation='swish'),
    Dense(20, activation='swish'),
    Dense(15, activation='swish'),
    Dense(1)
])

model.compile(
    optimizer=RMSprop(),
    loss=MeanSquaredError(),
    metrics=[RootMeanSquaredError()]
)
```

```python
history = model.fit(
    X_train, y_train,
    epochs=200,
    batch_size=64,
    validation_data=(X_test, y_test),
    verbose=1
)
```

```python
train_loss, train_rmse = model.evaluate(X_train, y_train, verbose=0)
test_loss, test_rmse = model.evaluate(X_test, y_test, verbose=0)

print(f"Training RMSE: {train_rmse:.2f}")
print(f"Testing RMSE: {test_rmse:.2f}")
```

```python
model.save('housing_price_ann.h5')

deployed_model = load_model('housing_price_ann.h5')
sample_input = np.expand_dims(X_test[0], axis=0)
predicted_price = deployed_model.predict(sample_input, verbose=0)
print(f"Predicted Price: {predicted_price[0][0]:.2f}")
```
